# Array operations


```python
# Arithmetic operations on numpy array are elementwise

# Let's create a couple of arrays
a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

# Array subtraction
c = a - b

# Let's look at the multiplication of the array
d = a*b


print(c)
print(d)
```

```python
# In Canada people use Celcius to measure temperature, conversly in the US farenheit 
# is used in place of it

farenheit = np.array([0, -5, -10, -15, 0])

# The formula for conversion to farenheit is ((*F - 32) * 5/9 = *C)
celcius = ((farenheit - 31) * 5/9)
celcius
```

```python
# Moreover we can get boolean asa an array elements
celcius > -20
```

```python
# Also, we can use modulus operator
celcius % 2  == 0
```

```python
# Let's look at the matrix product A*B, elementwise product
A = np.array([[1,2], [0, 1]])
B = np.array([[2,1], [1, 1]])

print(A*B)
```

```python
# For matrix product, we can use @ sign or dot function
print(A@B)
```

```python
# We can quickly see the shape of a matrix
A.shape
```

```python
# Upcasting

array1 = np.array([[1,2, 3], [3,4, 5]])
print(array1.dtype)
array2 = np.array([[1.1, 2.2, 3.4], [3.2, 3.4, 4.4]])
print(array2.dtype)
```

```python
# Let's do addition of two arrays
array3 = array1 + array2
print(array3)
print(array3.dtype)
```

```python
# Numpy arrays have some interesting functions sum(), max(), min(), mean()
print(array3.max())
print(array3.min())
print(array3.sum())
print(array3.mean())
```

```python
# Let's create an array with 15 elements, ranging from 1 to 15,
# with a dimension of 3x5
b = np.arange(1, 16, 1).reshape(3, 5)
print(b)
```

```python
# Image data array
from PIL import Image
from IPython.display import display

# Let's just look at the image
img = Image.open('img.png')
display(img)
```

```python
# Convert image to a numpy array
array = np.array(img)
print(array.shape)
array
```

```python
# uint means they are unsigned integers (so non negative numbers) and 8 means 8 bits per byte.
# This means each value can be up to 2*2*2*2*2*2*2*2=256 in size(actually 255, beacuse we start at zero)
# For black and white images black is stored as zero and white as 255. We invert this image

mask = np.full(array.shape, 255)
mask
```

```python
# Let's subtract that from the modified array
modified_array = array - mask

# Convert all values to positive values
modified_array = modified_array*-1

modified_array = modified_array.astype(np.uint8)
modified_array
```

```python
# Let's display this

display(Image.fromarray(modified_array))
```

```python
reshaped = np.reshape(modified_array, (583308))
```

```python
display(reshaped)
```

```python
display(Image.fromarray(reshaped))
```

# Indexing, slicing and iterating


```python
a = np.array([2,3,4,5])
a[2]
```

```python
a = np.array([[1,2], [3,4], [5,6]])
a
```

```python
# Access to multidimensional array elements
a[1,1]
```

```python
# If we want to get multiple elements
np.array([a[1,1],a[0,1], a[2,1]])
```

```python
# Another way called "zip"
print(a[[0,1,2], [0,1,1]])
```

# Boolean indexing


```python
print(a>5)
```

```python
# Get elemt from array given that the condition is true
print(a[a>5])
```

# Slicing


```python
a = np.array([1,2,3,4,5,6])
a[:3]
```

```python
print(a[2:4])
```

```python
# For multidimensional arrays it works similarly
a = np.array([[1,2,3,4], [4,5,6,7], [8,9,10,11]])
a
```

```python
a[:2]
```

```python
a[:2, 1:3]
```

```python
# Modifying a subarray will modify the original array
sub_array = a[:2, 1:3]
print(sub_array[0,0])
sub_array[0,0] = 50
print(sub_array[0,0])
print(a[0,1])
```

# Trying Numpy with Datasets



```python
# To load a dataset in Numpy, we can use the genfromtxt() function.
# filename, delimiter, number of rows to skip 

rice = np.array([[1,2,3,4], [2,3,4,5], [7,9,0,1],[2,3,5,7], [7,5,2,1], [3,4,6,8], [3,2,5,6], [3,4,5,8]])
print("one integer 0 for slicing: ", rice[:, 0])
print("0 to 1 for slicing: \n", rice[:, 0:1])
```

```python
print(rice[:, 0:3])
```

```python
rice[:, [0,3]]
```

```python
rice[:, -1].mean()
```
